# 3. Limpeza e Formatação dos Dados

**Dados nulos**

```python
df.isna().sum()
```

**Limpeza de Dados Categóricos**

```python
categorical_columns = ['Código do fundo', 'Setor']
```

```python
# Encontrar os valores que estão nulos na coluna 'Setor' e colocar na variável idx

idx = df[df['Setor'].isna()].index
idx
```

```python
# Excluir com .drop os valores nulos

df.drop(idx, inplace=True)
```

```python
df['Setor'].isna().sum()
```

```python
# Transformar as colunas categóricas do tipo 'object' para 'category'

df[categorical_columns] = df[categorical_columns].astype('category')
```

```python
df.info()
```

**Limpeza de Dados Numéricos**

```python
# subset das colunas númericas do tipo float com iloc

df.iloc[:,2:-1]
```

```python
# Colocar as colunas numéricas dentro de uma lista para usar como filtro do DataFrame

col_floats = list(df.iloc[:,2:-1].columns)
```

```python
# substituir dados nulos nas colunas numéricas com 0

df[col_floats] = df[col_floats].fillna(value=0)
```

```python
# Dados numéricos zerados

df[col_floats].isna().sum()
```

**Formatação das colunas de string para floats**

```python
# Aplicar uma função do DataFrame todo para retirar caracteres strings "R$", ".0" e "." 

df[col_floats].applymap(lambda x: str(x).replace('R$', '').replace('.0', '').replace('.', '').replace('%', '').replace(',', '.'))
```

```python
df[col_floats] = df[col_floats].applymap(lambda x: str(x).replace('R$', '').replace('.0', '').replace('.', '').replace('%', '').replace(',', '.'))
```

```python
# Transformar o tipo das colunas numéricas para float

df[col_floats] = df[col_floats].astype('float')
```

```python
df.describe()

# Agora as colunas podem ser calculadas, pois agora estão devidamente formatadas e limpas. 
# Mas as colunas P/VPA possuem valores infinitos e está em um escala diferente
```

**Formatação de valores infinitos**

```python
# Checar valores infinitos com Pandas

df[np.isinf(df[col_floats]).any(1)]
```

```python
# excluir o valor com número infinito 

drop_inf = df[np.isinf(df[col_floats]).any(1)].index
df.drop(drop_inf, inplace=True)
```

**Alteração de escala em uma coluna**

```python
df['P/VPA'] = df['P/VPA']/100
```

```python
# DataFrame final

df.describe()
```

# Criando uma função com uma estratégia para oportunidades do mercado

```python
def oportunidade_media_setor(df, setor='Shoppings', label_setor='Setor'):
    
    media_setor = df_aux.groupby('Setor').agg(["mean", "std"])
    
    df_setor = df[df[label_setor].isin([setor])]
    
    filter_ = (df_setor['Quantidade Ativos'] > 5) &\
              (df_setor['Liquidez Diária'] > 5000) &\
              (df_setor['P/VPA'] < 1.0) &\
              (df_setor['DY (12M) Acumulado'] > media_setor.loc[setor, ('DY (12M) Acumulado', 'mean')])
    
    print('média do setor Yield: {}'.format(media_setor.loc[setor, ('DY (12M) Acumulado', 'mean')]))
    print('média do setor P/VPA: {}'.format(media_setor.loc[setor, ('P/VPA', 'mean')]))
    print('média do setor Ativos: {}'.format(media_setor.loc[setor, ('Quantidade Ativos', 'mean')]))
    
    return df_setor[filter_]
```

```python
list(df['Setor'].unique())
```

```python
oportunidade = oportunidade_media_setor(df_aux, setor='Shoppings')
oportunidade.sort_values('DY (12M) Acumulado', ascending = False, inplace=True)
oportunidade
```
